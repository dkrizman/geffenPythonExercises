Bubble sort

```python
l = [rd.randint(1,200) for i in range(2000)]

def bubbleSort(l):
    if len(l) in [0,1]:
        return l
    else:
        for i in range(1,len(l)):
            for j in range(len(l)):
                if l[j] > l[i]:
                    temp = l[i]
                    l[i] = l[j]
                    l[j] = temp
    return l

bubbleSort(l)
```

merge Sort

```python
def merge(l,r):
    res = []
    while len(l)!=0 and len(r)!=0:
        if l[0]<r[0]:
            res.append(l[0])
            l.remove(l[0])
        else:
            res.append(r[0])
            r.remove(r[0])
    if len(l) == 0:
        res = res + r
    else:
        res = res + l
    return res


def mergeSort(ul):
    if len(ul) <=1:
        return ul
    
    middle = len(ul)//2
    l = ul[:middle]
    r = ul[middle:]
    
    l = mergeSort(l)
    r = mergeSort(r)
    return merge(l,r)

# l = [100,23,5,4,23,34,12,2,4,1]
l = [rd.randint(1,200) for i in range(2000)]
mergeSort(l)
```

insertion sort

```python
def insertionSort(l):
    for i in range(1,len(l)):
        j=i-1
        nextElem = l[i]
        while l[j]>nextElem and j>=0:
            l[j+1] = l[j]
            j=j-1
        l[j+1] = nextElem
    return l

l = [rd.randint(1,200) for i in range(2000)]
insertionSort(l)    
```
